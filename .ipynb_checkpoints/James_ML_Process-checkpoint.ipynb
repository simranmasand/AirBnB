# Creating A RandomForrest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('nhanes_cleaned_01feb_2.csv')
```

```python
class_df = df.copy()
```

```python
class_df.drop(columns = ['glycatedHb%',
'insulinumol',
'HbA1c_correct',
'HbA1c',
 'Anaemia',
 'T1DM',
 'insulin_prescribed',
 'bad_hba1c',
 'macrovascular_composite',
 'Doctor told you have diabetes',
 'Ever told you have prediabetes',
 'Ever told have health risk for diabetes',
 'Had blood tested past three years',
 'Taking insulin now',
 'How often check blood for glucose/sugar',
 'Past year times Dr check feet for sores',
 'How often do you check your feet', 'Unnamed: 0'],
       
inplace= True)
```

```python
class_df = class_df.dropna(thresh=df.shape[0]*0.6,how='all',axis=1)
class_df = class_df.dropna()
```

```python
y = class_df.pop('Known_T2DM')
```

```python
X = class_df
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
######
X_train = pd.DataFrame(X_train_std, columns = X_train.columns)
X_test = pd.DataFrame(X_test_std, columns = X_test.columns)
```

```python
clf = RandomForestClassifier(n_estimators = 100)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
clf.score(X_train, y_train)
```

```python
cross_validated_scores = np.mean(cross_val_score(clf, X_test, y_test, cv=5))
```

```python
cross_validated_scores
```

```python
X
```

```python
from sklearn import metrics
from sklearn.metrics import classification_report
```

```python
X
```

```python
y_pred = clf.predict(X_test)
y_true = y_test
```

```python
print(classification_report(y_true, y_pred, 
                            target_names=['class 0', 'class_1']
                           ))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_true, y_pred)
```
